```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("12_irish.csv",sep=';')  
```

```python
def get_avg_feature(data, indexes):
    avg_col = list()
    for index in indexes:
        avg_col.append(data[data.keys()[index]])
    return np.mean(avg_col, axis=0)
```

```python
def get_random_feature(data, lb, ub):
    return np.random.uniform(low=lb, high=ub, size=(len(data),))
```

```python
def add_noise(data, index, lb, ub):
    noise = np.random.uniform(low=lb, high=ub, size=(len(data),))
    return data[data.keys()[index]] + noise
```

```python
#avg_sepal = get_avg_feature(data, [0,1])                
random_feature1 = get_random_feature(data, 0, 2)
#random_noise = add_noise(data, 0, 0, 1)
random_feature2 = get_random_feature(data, 1,2 )
random_feature3 = get_random_feature(data, 1, 2)
random_feature4 = get_random_feature(data, -1, 1)
random_feature5 = get_random_feature(data, -1, 0)
```

```python
#data.insert(4, 'avg_sepal', avg_sepal)
data.insert(0, 'randm_f1', random_feature1)
data.insert(0, 'randm_f2', random_feature2)
data.insert(0, 'randm_f3', random_feature3)
data.insert(0, 'randm_f4', random_feature4)
data.insert(0, 'randm_f5', random_feature5)

#data.insert(4, 'sepal_length_noise', random_noise)
#data.insert(4, 'randm_f1', random_feature)
```

```python
def get_sum_feature(data, indexes):
    avg_col = list()
    for index in indexes:
        avg_col.append(data[data.keys()[index]])

    return( np.sum(avg_col, axis=0))


```

```python
sum_f1 = get_sum_feature(data, [0,1,2,3,4])                     
avg_f2 = get_avg_feature(data, [5,6,7,8,9])
sum_f3 = get_sum_feature(data, [10,11,12,13])
avg_f4 = get_avg_feature(data, [13,14])
sum_f5 = get_sum_feature(data, [15,16])



```

```python
data.insert(0, 'f1', sum_f1)
data.insert(0, 'f2', avg_f2)
data.insert(0, 'f3', sum_f3)
data.insert(0, 'f4', avg_f4)
data.insert(0, 'f5', sum_f5)
```
